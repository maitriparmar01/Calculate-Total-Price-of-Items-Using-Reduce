#### Calculate Total Price of Items Using Reduce (Project 1)

- It repeatedly adds each item price in a list and returns the final total amount.

```python
Items = ['Apple','Perfume','T-shirt','Bag']
Prices = [500,2000,3000,2800]
```

```python
from functools import reduce

Total = reduce(lambda x,y:x+y,Prices)
```

```python
print(f'Total Price for this items: {items} : {Total}')
```
